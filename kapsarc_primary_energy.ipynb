```python
df = pd.read_csv('primary-energy-consumption-1965-2015.csv',sep=';')
df.head(10)
```

```python
df['Country'].unique()
```

```python
pd.DataFrame(df['Country'].unique())
```

```python
df_us = df[df['Country']=='US'].loc[:,['Year','Energy Consumption']]
print(df_us)
df_us.info()
```

```python
x=df_us
x=x.sort_values(by='Year')
x=x.reset_index(drop=True)
x
```

```python
x.plot(x='Year')
plt.title('primary-energy-consumption-US-1965-2015')
plt.ylabel('Mtoe million tonnes of oil equivalent')
```
