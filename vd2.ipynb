# Total energy consumption

link
https://www.eia.gov/opendata/qb.php?category=40236

```python
# EIA Data Sets > State Energy Data System (SEDS) > Consumption > Total All Sectors
url = 'http://api.eia.gov/category/?api_key=ff6b6c88bf6550d3c017016868cc1045&category_id=40236'
req = requests.get(url)
data = req.json()
data
```

```python
pd.DataFrame(data)
```

```python
all_states=pd.DataFrame(data['category']['childseries'])
all_states.info()
```

```python
all_states
```

```python
# for example: Alabama
url_AL = 'http://api.eia.gov/series/?api_key=ff6b6c88bf6550d3c017016868cc1045&series_id=SEDS.TETCB.AL.A'
req_al = requests.get(url_AL)
data_al = req_al.json()
data_al
```

```python
pd.DataFrame(data_al['series'][0]['data']).head()
```

```python
alabama = pd.DataFrame(data_al['series'][0]['data'])
alabama.columns = ['year','total']
alabama=alabama.sort_values(by=['year'])
alabama.head()
```

```python
alabama.plot(x='year')
plt.ylabel('trillion BTU')
plt.title('Alabama')
```
